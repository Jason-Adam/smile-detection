# Automated Smile Detection Part 3: Testing the Model  
This section will serve as a place to test the model on random smile images  

# Imports

```python
import os
import warnings

import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

warnings.simplefilter("ignore")
```

# Load Photos

```python
smile_test = cv2.imread("smile_test.jpg", cv2.IMREAD_UNCHANGED)
no_smile_test = cv2.imread("no_smile_test.jpg", cv2.IMREAD_UNCHANGED)
```

```python
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(10, 5))

ax[0].imshow(smile_test)
ax[1].imshow(no_smile_test)

plt.show()
```

```python
processed_smile_test = cv2.resize(smile_test, (150, 150)) * (1.0 / 255)
processed_smile_test = np.expand_dims(processed_smile_test, axis=0)

processed_no_smile_test = cv2.resize(no_smile_test, (150, 150)) * (1.0 / 255)
processed_no_smile_test = np.expand_dims(processed_no_smile_test, axis=0)
```

# Load Model & Test

```python
loaded_model = tf.keras.models.load_model("../models/smile_model.h5")
loaded_model.layers[0].input_shape

warnings.simplefilter("ignore")
```

```python
smile_result = loaded_model.predict_classes(processed_smile_test)
no_smile_result = loaded_model.predict_classes(processed_no_smile_test)

print(smile_result[0][0])
print(no_smile_result[0][0])
```

We can see from the predictions that my model isn't performing well and isn't picking up my smile photo.
